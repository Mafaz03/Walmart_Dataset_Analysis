```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch import nn
```

```python
from tqdm.notebook import tqdm
```

### Reading csv files

```python
X_train = pd.read_csv('train.csv')
```

```python
X_train['Year'] = pd.DatetimeIndex(X_train['Date']).year 
X_train['month'] = pd.DatetimeIndex(X_train['Date']).month
```

## Splitting Data to Train and Test

```python
y_train_data = X_train['Weekly_Sales'][ : 10000]
y_test_data = X_train['Weekly_Sales'][10000 : 20000]
```

```python
X_train_data = X_train['month'][ : 10000]
X_test_data = X_train['month'][10000 : 20000]
```

### Function for plotting graph

```python
def ploting(train_data = X_train_data,
            train_labels = y_train_data,
            test_data = X_test_data,
            test_labels = y_test_data,
            predictions = None):

  plt.figure(figsize=(10 , 10))

  plt.scatter(train_data , train_labels ,c='b' , s = 4 , label = "Trainging Data")
  plt.scatter(test_data , test_labels ,c='g' , s = 4 , label = "Testing Data")

  if predictions is not None:
    plt.scatter(test_data , predictions ,s = 5 ,c = 'r' , label = "Predictions")

  plt.legend(prop = {'size' : 14})
```

## Training Loop

```python
class LRM(nn.Module):
  def __init__(self):
    super().__init__()
    self.weight = nn.Parameter(torch.randn( 1 , requires_grad = True , dtype = torch.float))
    self.bias = nn.Parameter(torch.randn( 1 , requires_grad = True , dtype = torch.float))
    
  def forward(self , x:torch.Tensor) -> torch.Tensor:
    return self.weight * x + self.bias
```

```python
Walmart_model = LRM()
```

```python
Walmart_model.parameters()
```

```python
list(Walmart_model.parameters())
```

```python
type(y_test_data)
```

### Pandas Series to Tensors 

```python
def df_to_tensor(df):
    return torch.from_numpy(df.values).float()
```

```python
y_train_data = df_to_tensor(y_train_data)
y_test_data = df_to_tensor(y_test_data)
X_train_data = df_to_tensor(X_train_data)
X_test_data = df_to_tensor(y_test_data)
```

```python
type(y_test_data) , type(X_train_data) , type(y_train_data) , type(y_train_data)
```

```python
with torch.inference_mode():
    y_prediction = Walmart_model(y_test_data)
y_prediction
```

```python
Walmart_model.state_dict()
```

### Prediction Before Training

```python
ploting(predictions = y_prediction)
```

### Setting up Loss and optimizers

```python
loss_func = nn.L1Loss()

optimizer = torch.optim.SGD(params = Walmart_model.parameters() , lr = 1)
```

# Training

```python
epochs = 100000

epoch_list = []
loss_list = []
test_loss_list = []

for epoch in range(epochs):
    
    Walmart_model.train()
    y_prediction = Walmart_model(X_train_data)
    loss = loss_func(y_prediction , y_train_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    Walmart_model.eval()
    with torch.inference_mode():
        test_prediction = Walmart_model(X_train_data)
        test_loss = loss_func(test_prediction , y_test_data)
        
        if epoch % 10000 == 0:
            print(f'epoch : {epoch}     loss : {loss}     test_loss : {test_loss}')
            epoch_list.append(epoch)
            loss_list.append(loss)
            test_loss_list.append(test_loss)
    
```

# Loss Curves

```python
for i in range(len(loss_list)):
    loss_list[i] = loss_list[i].item()
```

```python
for i in range(len(test_loss_list)):
    test_loss_list[i] = test_loss_list[i].item()
```
